```python
import cv2 as cv2
import os
import matplotlib.pyplot as plt

for gtFiles in os.listdir("/content/drive/MyDrive/Fall 2022/CV/Shadow Removal DDPM Project/ISTD_Dataset/test/test_A/"):
  count = 3
  if ("-v" in gtFiles or "-h" in gtFiles):
    continue
  img_gt = cv2.imread("/content/drive/MyDrive/Fall 2022/CV/Shadow Removal DDPM Project/ISTD_Dataset/test/test_A/" + gtFiles)
  img_samples = []
  gtFiles = gtFiles.split(".")
  for files in os.listdir("/content/drive/MyDrive/Fall 2022/CV/Shadow Removal DDPM Project/ilvr_adm/output/shadow/"):
    if (count <= 0):
      break;
    if (gtFiles[0] in files):
      print(gtFiles[0])
      print(files)
      count = count - 1
      img_sample = cv2.imread("/content/drive/MyDrive/Fall 2022/CV/Shadow Removal DDPM Project/ilvr_adm/output/shadow/" + files)
      img_samples.append(img_sample)
  
  if (len(img_samples) < 3):
    continue
  
  plt.figure(figsize=(15, 5))
  plt.subplot(1, 4, 1)
  plt.imshow(img_gt)
  plt.title('Reference Image')
  plt.axis("off")

  plt.subplot(1, 4, 2)
  plt.imshow(img_samples[0])
  plt.title('Sample 1')
  plt.axis("off")

  plt.subplot(1, 4, 3)
  plt.imshow(img_samples[1])
  plt.title('Sample 2')
  plt.axis("off")

  plt.subplot(1, 4, 4)
  plt.imshow(img_samples[2])
  plt.title('Sample 3')
  plt.axis("off")

  plt.show()
```
