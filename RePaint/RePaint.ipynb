```python
import cv2 as cv2
import os
import matplotlib.pyplot as plt

for files in os.listdir("/content/drive/MyDrive/Fall 2022/CV/Shadow Removal DDPM Project/RePaint/log/face_example/gt1"):
  img_gt = cv2.imread("/content/drive/MyDrive/Fall 2022/CV/Shadow Removal DDPM Project/RePaint/log/face_example/gt1/" + files)
  img_keep_mask = cv2.imread("/content/drive/MyDrive/Fall 2022/CV/Shadow Removal DDPM Project/RePaint/log/face_example/gt_keep_mask1/" + files)
  img_masked = cv2.imread("/content/drive/MyDrive/Fall 2022/CV/Shadow Removal DDPM Project/RePaint/log/face_example/gt_masked1/" + files)
  img_inpainted = cv2.imread("/content/drive/MyDrive/Fall 2022/CV/Shadow Removal DDPM Project/RePaint/log/face_example/inpainted1/" + files)
  
  plt.figure(figsize=(15, 5))
  plt.subplot(1, 4, 1)
  plt.imshow(img_gt)
  plt.title('Ground Truth')
  plt.axis("off")

  plt.subplot(1, 4, 2)
  plt.imshow(img_keep_mask)
  plt.title('Shadow Mask')
  plt.axis("off")

  plt.subplot(1, 4, 3)
  plt.imshow(img_masked)
  plt.title('Masked Image')
  plt.axis("off")

  plt.subplot(1, 4, 4)
  plt.imshow(img_inpainted)
  plt.title('Shadow Free Image')
  plt.axis("off")

  plt.show()
```
